```python
import pandas as pd
import os
import numpy as np

# Pfad zum Ordner setzen, der die Textdateien enthält
ordner_pfad = 'C:/Users/erikm/Desktop/Diplomarbeit Erik Marr/Daten/DatenSimulation'  # Ersetzen Sie dies mit dem tatsächlichen Pfad

# Liste, um die DataFrames zu speichern
df_list = pd.DataFrame()

df_x = []
df_y = []
df_t = []


df_x = pd.read_csv("C:/Users/erikm/Desktop/Diplomarbeit Erik Marr/Daten/DatenSimulation/Skalen/TPath_coorX.txt",delimiter='  ', engine='python' )
df_y = pd.read_csv("C:/Users/erikm/Desktop/Diplomarbeit Erik Marr/Daten/DatenSimulation/Skalen/TPath_coorY.txt",delimiter='  ', engine='python' )
df_t = pd.read_csv("C:/Users/erikm/Desktop/Diplomarbeit Erik Marr/Daten/DatenSimulation/Skalen/TPath_time.txt",delimiter='  ', engine='python' )
df_x_list = df_x.iloc[:, 0].tolist()
df_y_list = df_y.iloc[:, 0].tolist()
df_t_list = df_t.iloc[:, 0].tolist()
```

```python
# Pfad zum Verzeichnis mit den CSV-Dateien
verzeichnis = ordner_pfad

# Durchlaufen aller Dateien im Verzeichnis
for dateiname in os.listdir(verzeichnis):
    if dateiname.endswith('.txt'):
        # Pfad zur aktuellen CSV-Datei
        voller_pfad = os.path.join(verzeichnis, dateiname)

        # Einlesen der CSV-Datei in ein DataFrame
        df = pd.read_csv(voller_pfad, delimiter='  ', engine='python')
        df.loc[-1] = df_y_list
        df.index = df.index + 1
        df = df.sort_index()

        # Transponieren des DataFrame
        df_transponiert = df.T

        # Bestimmen des neuen Dateinamens
        neuer_dateiname = 'transponiert_' + dateiname
        neuer_voller_pfad = os.path.join(verzeichnis, neuer_dateiname)

        # Speichern des transponierten DataFrame in einer neuen CSV-Datei
        df_transponiert.to_csv(neuer_voller_pfad, header=False)

print("Transposition aller CSV-Dateien abgeschlossen.")
```
