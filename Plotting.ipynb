```python
data = np.array([0.0017, 0.0012, 0.002, 0.0014, 0.002, 0.0014, 0.0066, 0.0049])
categories = ['D1', 'D2', 'D3', 'D4']

fig, ax = plt.subplots()
width = 0.2  # Width of the bars
positions = np.arange(len(categories))  # Base positions for each category

# Plotting two bars for each category with custom labels
ax.bar(positions - width/2, data[::2], width, color='red', label='Test_MSE')
ax.bar(positions + width/2, data[1::2], width, color='blue', label='Test_MAE')

ax.set_ylabel('Absoluter Wert')
ax.set_title('MSE und MAE f√ºr unterschiedliche Aufl√∂sungen')
ax.set_xticks(positions)
ax.set_xticklabels(categories)

# Manually adjust the legend
ax.legend(title='Legende')

plt.grid(axis='y')
plt.savefig('C:/Users/erikm/Desktop/Diplomarbeit Erik Marr/Bilder Diplomarbeit/Plotting/MSE und MAE_Aufl√∂sung')
plt.show()
```
