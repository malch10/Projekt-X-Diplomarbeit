```python
import numpy as np
import matplotlib.pyplot as plt

# Original data and categories
data = np.array([7.41, 7.57, 7.06, 9.58, 6.98, 7.9])
categories = ['1', '9', '21', '1', '9', '21']

# Numerische Indizes für die x-Positionen verwenden
x_positions1 = np.arange(len(categories[:3]))
x_positions2 = np.arange(len(categories[3:]))

# Max-Wert Array
max_data = 10

# Erstellen von Subplots
fig, (ax1, ax2) = plt.subplots(1, 2)

# Erstes Diagramm
ax1.bar(x_positions1, data[:3], color='blue', width=0.3, align='center')
ax1.set_xticks(x_positions1)
ax1.set_xticklabels(categories[:3], ha='center')
ax1.set_ylim(0, max_data + 1)  # Ein wenig mehr Platz am oberen Rand
ax1.set_ylabel('MAE in K')
ax1.set_title('High')
ax1.grid(axis='y')

# Zweites Diagramm
ax2.bar(x_positions2, data[3:], color='blue', width=0.3, align='center')
ax2.set_xticks(x_positions2)
ax2.set_xticklabels(categories[3:], ha='center')
ax2.set_title('Low')
ax2.set_ylim(0, max_data + 1)  # Einheitliche y-Achse
ax2.yaxis.set_tick_params(labelleft=False)
ax2.grid(axis='y')

ax1.set_xlim(-0.5, len(categories[:3]) - 0.5)
ax2.set_xlim(-0.5, len(categories[3:]) - 0.5)

plt.tight_layout()
plt.xlabel('Anzahl an betrachteten Zeitschritten im Datenpaket', ha='right')
plt.savefig('C:/Users/erikm/Desktop/Diplomarbeit Erik Marr/Bilder Diplomarbeit/Plotting/MAE_D3-D4_ALL.pdf', format='pdf') 

plt.show()
```
