```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([7.41, 7.57, 7.06, 9.58, 6.98, 7.9])
categories = ['D3_I_F_1', 'D3_t_9_I_F_1', 'D3_t_21_I_F_2', 'D4_I_F_2', 'D4_t_9_I_F_2', 'D4_t_21_I_F_1']

# Check if the length of data matches the length of categories
if len(data) != len(categories):
    raise ValueError("Length of 'data' and 'categories' should be the same.")

# Plotting MAE values as a bar plot
fig, ax = plt.subplots()
ax.bar(categories, data, color='blue', width=0.4)  # Adjusted width for better visualization

ax.set_ylabel('MAE in K')
ax.set_title('MAE nach Datenpaket')
ax.set_xticklabels(categories, rotation=45, ha='right')  # Adjusted rotation and alignment for better readability

plt.grid(axis='y')
plt.tight_layout()  # Ensures tight layout to prevent clipping of labels
plt.savefig('C:/Users/erikm/Desktop/Diplomarbeit Erik Marr/Bilder Diplomarbeit/Plotting/MAE_D3-D4_ALL.png', dpi=300)  # Save as PNG with higher resolution
plt.show()
```
